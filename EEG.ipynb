```julia
using PyPlot
using DataFrames
using Dates
import CSV
include("jules.jl")
```

```julia
function plotAverage(timestamps, data)
    smoothing = 19;
    
    average_delta = (data[:,:Delta_TP9] .+ data[:,:Delta_TP10] .+ data[:, :Delta_AF7] .+ data[:, :Delta_AF8])./4.0
    average_theta = (data[:,:Theta_TP9] .+ data[:,:Theta_TP10] .+ data[:, :Theta_AF7] .+ data[:, :Theta_AF8])./4.0
    average_gamma = (data[:,:Gamma_TP9] .+ data[:,:Gamma_TP10] .+ data[:, :Gamma_AF7] .+ data[:, :Gamma_AF8])./4.0
    average_alpha = (data[:,:Alpha_TP9] .+ data[:,:Alpha_TP10] .+ data[:, :Alpha_AF7] .+ data[:, :Alpha_AF8])./4.0
    average_beta  = (data[:,:Beta_TP9]  .+ data[:,:Beta_TP10]  .+ data[:, :Beta_AF7]  .+ data[:, :Beta_AF8])./4.0
    
    plot(timestamps, Jules.smooth(average_delta, smoothing), label="Delta")
    plot(timestamps, Jules.smooth(average_theta, smoothing), label="Theta")
    plot(timestamps, Jules.smooth(average_gamma, smoothing), label="Gamma")
    plot(timestamps, Jules.smooth(average_alpha, smoothing), label="Alpha")
    plot(timestamps, Jules.smooth(average_beta,  smoothing), label="Beta")
    legend()
end
```

```julia
meditation_data = CSV.read("mindMonitor_2021-10-30--15-14-25.csv", DataFrame);
improv_data = CSV.read("mindMonitor_2021-10-30--14-49-38.csv", DataFrame);
arabesque_data = CSV.read("mindMonitor_2021-10-30--14-36-38.csv", DataFrame);
arabesque2_data = CSV.read("mindMonitor_2021-10-30--14-53-46.csv", DataFrame);
```

```julia
function cleanData(data)
    return data[completecases(data[:,1:end-1]),:];
end

function makeTimestamps(data)
    timestamps = map(x -> floor(datetime2unix(DateTime(x, dateformat"y-m-d H:M:S.s"))), data[:,:TimeStamp]);
    timestamps = (timestamps .- timestamps[1])./60;
    return timestamps;
end

cleaned_meditation_data = cleanData(meditation_data);
meditation_timestamps = makeTimestamps(cleaned_meditation_data);

cleaned_improv_data = cleanData(improv_data);
improv_timestamps = makeTimestamps(cleaned_improv_data);

cleaned_arabesque_data = cleanData(arabesque_data);
arabesque_timestamps = makeTimestamps(cleaned_arabesque_data);

cleaned_arabesque2_data = cleanData(arabesque2_data);
arabesque2_timestamps = makeTimestamps(cleaned_arabesque2_data);
```

```julia
plotAverage(meditation_timestamps, cleaned_meditation_data)
axvline(x=0.5)
axvline(x=3.5)
axvline(x=6.5)
axvline(x=9.5)
xlabel("Time (min)")
ylabel("Signal (dB)")
```

```julia
plotAverage(arabesque_timestamps, cleaned_arabesque_data);
```

```julia
plotAverage(improv_timestamps, cleaned_improv_data);
```

```julia
plotAverage(arabesque2_timestamps, cleaned_arabesque2_data)
```
